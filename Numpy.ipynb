### Changing the shape of an array (reshape)

```python
arr4= np.random.randint(10,100,100)
arr4.reshape(10,10)
#  While reshapingthe array,no of elements in the original array and multiplication of dimension on reshaping must be equal
```

```python
arr4
```

```python
# reshaping with -ve numbers
#  only one dim can be -ve 
#  numpy automatically calculates the other dim if we specify our req dim

arr4.reshape(5,-1) # now it reshaped it to 5 rows and 25 cols
```

```python
arr4.reshape(-1,25)
```

### Iterating over an array

```python
import numpy as np
```

```python
arr7= np.random.randint(10,20,30).reshape(5,6)
arr7
```

```python
for i in arr7:
    print(i)
```

```python
for i in arr7:
    for j in i:
        print(j, end=" ")
```

```python
for i in np.nditer(arr7, order='C'):  # unpacks in Row wise format
    print(i,end=" ")
```

```python
for i in np.nditer(arr7, order='F'):  # unpacks in coloumn wise format
    print(i,end=" ")
```
