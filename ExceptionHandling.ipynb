# Exception Handling 

## 1. Compile time
## 2. Runtime

 try, except, else, finally, raise

- try — We write the piece of code which is suppose to raise the error
- except— The code which you want to execute if error is raised
- else — The code which you want to execute if error is not raised
- finally — The code which you want to execute irrespective of the errors
- raise - used to raise custom exception

```python
a= int(input("enter a"))
b=  int(input("enter b"))
try:
    res= a/b
except:
    print("some thing went wrong")
else:
    print(res)
finally:
    print(a)
    print(b)
```

### User defined exceptions

```python
class MyException(Exception):
    def __init__(self,msg):
        self.msg= msg
        
a= 10
b= int(input())
try:
    if b==0:
        raise MyException("can't divide by zero")
except Exception as e:
    print(e)
else:
    print(a/b)
```

## Pickle module
works with binary files

- Dump => inserts data into file in binary format
- Load => coverts the binary data to readable format

```python
import pickle

def createFile():
    no_of_entries= int(input("no of books"))
    with open('Book.dat','wb') as f1:
        for i in range(no_of_entries):
            pickle.dump([input("id"),input("name"),input("author"),input("price")],f1)
#            pickle.dump function inserts the data into file in binary format 
            
createFile()
```
