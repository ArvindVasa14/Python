```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distributions
mean = 0
variance_1 = 1  # Smaller variance
variance_2 = 4  # Larger variance

# Generate x-values for the curve
x = np.linspace(-10, 10, 100)

# Calculate the probability density function (PDF) for the two normal distributions
pdf_1 = 1 / np.sqrt(2 * np.pi * variance_1) * np.exp(-0.5 * ((x - mean) ** 2) / variance_1)
pdf_2 = 1 / np.sqrt(2 * np.pi * variance_2) * np.exp(-0.5 * ((x - mean) ** 2) / variance_2)

# Plotting the normal distribution curves
plt.plot(x, pdf_1, label='Smaller Variance')
plt.plot(x, pdf_2, label='Larger Variance')

# Set plot labels and title
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution Curves')

# Set legend
plt.legend()

# Display the plot
plt.show()
```
