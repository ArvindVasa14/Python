# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr1= np.arange(1,10,1)
arr1
```

```python
arr2= arr1**2
arr2
```

```python
plt.plot(arr1,arr2)
plt.xlabel('Arr1')
plt.ylabel('Arr2')
plt.title('Arr1 vs Arr2')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(arr1,arr2)
plt.subplot(1,2,2)
plt.plot(arr2,arr1)

# displaying multiple plots in same canvas
```

```python
plt.subplot(2,2,1)
plt.plot(arr1,arr2)
plt.subplot(2,2,2)
plt.plot(arr2,arr1)
plt.subplot(2,2,3)
plt.plot(arr2,arr1**3)
```

### Object oriented Approach

```python
# creating a figure
fig= plt.figure()

# adding the axes
ax= fig.add_axes([0.1,0.2,0.8,0.8])

# plotting the graph
ax.plot(arr1,arr2)

ax.set_xlabel('Arr1')
ax.set_ylabel('Arr2')
ax.set_title('Arr1 vs Arr2')
```

###  creating inset graph

```python
# creating a figure
fig= plt.figure()

# adding the axes
ax1= fig.add_axes([0.1,0.2,0.8,0.8])
ax2= fig.add_axes([0.2,0.65,0.3,0.3])

# plotting the graph
ax1.plot(arr1,arr2)
ax2.plot(arr2,arr1)

# label setting
ax1.set_xlabel('Arr1')
ax1.set_ylabel('Arr2')
ax1.set_title('Arr1 vs Arr2')

ax2.set_xlabel('Arr2')
ax2.set_ylabel('Arr1')
ax2.set_title('Arr2 vs Arr1')
```

###    plt.subplots() - Automatic Axis manager

```python
fig, ax= plt.subplots()
ax.plot(arr1,arr2)
```

```python
# Creating multiple plots using subplots
fig, ax= plt.subplots(nrows=2, ncols= 2) 
ax
```

```python
#  we use indexing for ploting in sub plot
fig, ax= plt.subplots(nrows=2, ncols= 2) 
ax[0][0].plot(arr1,arr2)
ax[1][1].plot(arr2,arr1)
plt.tight_layout()
```

### Univariate analysis

Univariate analysis refers to the analysis of a single variable in isolation, focusing on understanding its characteristics, distribution, and summary statistics. It involves examining and summarizing the data of a single variable without considering the relationships or interactions with other variables.

The main objective of univariate analysis is to gain insights into the specific variable under investigation. It aims to answer questions such as:

- What are the central tendency measures of the variable? (e.g., mean, median, mode)
- How is the variable distributed? (e.g., normal distribution, skewed distribution)
- What are the measures of dispersion or spread? (e.g., range, variance, standard deviation)
- Are there any outliers or extreme values?
- What are the summary statistics, such as minimum and maximum values, quartiles, and percentiles?
- Univariate analysis typically involves visualizing the data using graphical techniques such as histograms, bar plots, box      plots, and density plots. These visualizations help understand the distribution, identify patterns, detect outliers, and assess the skewness or symmetry of the variable.

Additionally, univariate analysis may include calculation of summary statistics, exploratory data analysis, and hypothesis testing specific to the single variable of interest.

Univariate analysis provides a foundational understanding of a variable's characteristics and forms the basis for further analysis, including bivariate and multivariate analysis, which involve examining the relationships between multiple variables.
