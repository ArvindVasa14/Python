### Array Broad Casting

```python
# arr1= [10,20,30,40]
# arr1 * 5 = [10,20,30,40] * [5,5,5,5]
# same with 2D arrays
# [[10,20,30,40],   [[5,5,5,5]
#  [10,20,30,40],    [5,5,5,5]
#  [10,20,30,40]]    [5,5,5,5]]

#  we can do brodcasting only if it is 1 X M

arr1= np.arange(10,15)
arr1
```

```python
arr2= np.random.randint(10,20,15).reshape(3,5)
arr2
```

```python
arr2 * arr1
# it brodcasts the array1 in below format
# [[10, 12, 10, 15, 11],  [[10, 11, 12, 13, 14]
#  [12, 13, 19, 14, 19],   [10, 11, 12, 13, 14]
#  [10, 11, 16, 13, 11]]   [10, 11, 12, 13, 14]]
```

 __For broadcasting to happen,__
 1. Either of the dimension should be same in both the cases
 2. The dimension which is not similar should be 1 for any one array
 3. or, It is also possible if dimension of the arrays are (1xn) and (nx1)

```python
arr3= np.random.randint(10,20,10).reshape(2,5) 
arr2*arr3

#  we cannot broadcast if any one array doesn't have dimension to be 1 in both array 
```

```python
arr4= np.random.randint(10,20,5).reshape(5,1)
arr4
```

```python
arr1
```

```python
arr4 * arr1 
# both the arrays are broadcasted into 5 X 5
```

## Ioc

In the context of the Pandas library in Python, the .loc indexer is used to access and modify data in a DataFrame using labeled indexes. It allows you to select data by specifying row and column labels.

Here are some common use cases for .loc:

- Selecting rows: You can use .loc to select specific rows from a DataFrame based on their labels. For example, df.loc[3] selects the row with the label "3" (assuming the index labels are integers).

- Slicing rows: .loc can be used to select a range of rows using label-based slicing. For example, df.loc[2:5] selects rows with labels from "2" to "5" inclusive.

- Selecting rows and columns: You can use .loc to select specific rows and columns simultaneously. For example, df.loc[3, 'column_name'] selects the value in the row labeled "3" and the column labeled 'column_name'. You can also pass lists of labels to select multiple rows or columns.

- Conditional selection: .loc can be combined with boolean conditions to select rows based on specific criteria. For example, df.loc[df['column_name'] > 5] selects all rows where the values in 'column_name' are greater than 5.

- Modifying values: You can assign new values to specific rows and columns using .loc. For example, df.loc[3, 'column_name'] = 10 modifies the value in the row labeled "3" and the column labeled 'column_name' to 10.

- It's important to note that the labels used with .loc are inclusive, meaning both the start and end labels are included in the selection. Also, .loc is primarily used for label-based indexing, whereas .iloc is used for integer-based indexing.

### Concatenation and Merging of a dataFrame

Concatenation combines the both dataFrames into one frame by appending the 2nd df to 1st df, if exists any dis-similar columns it fills them with NaN values
where as merge merges them based on there index and adds dis-similar columns to 1st df
